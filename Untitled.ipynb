```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
from sklearn.metrics import mean_squared_error, r2_score
```

## Load Dataset

```python
def load_wine_data(file_path):
    data = pd.read_csv(file_path, sep=';')
    return data


def create_class_labels(data):
    data['class'] = np.where(data['quality'] >= 6, 1, 0)
    return data


def split_and_scale_features(features, labels, test_size=0.2):
    X_train, X_test, y_train, y_test = train_test_split(
        features, labels, test_size=test_size, random_state=42
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, y_train, y_test
```

## kNN training Function

```python
def train_knn_model(X_train, y_train, k):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    return model
```

## Metrics Functon (A1)

```python
def classification_metrics(model, X, y_true):
    y_pred = model.predict(X)
    
    cm = confusion_matrix(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    return cm, precision, recall, f1
```

## Regression Metrics Function (A2)

```python
def regression_metrics(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)
    
    return mse, rmse, mape, r2
```

## Random 2D Data Generation (A3)

```python
def generate_training_points(n=20):
    X = np.random.uniform(1, 10, (n, 2))
    y = np.where(X[:, 0] + X[:, 1] > 10, 1, 0)
    return X, y


def plot_training_data(X, y):
    plt.scatter(X[y==0][:,0], X[y==0][:,1], color='blue', label='Class 0')
    plt.scatter(X[y==1][:,0], X[y==1][:,1], color='red', label='Class 1')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.title("Training Data")
    plt.show()
```

## Decision Space Plotting (A4 & A5)

```python
def generate_test_grid():
    x_vals = np.arange(0, 10, 0.1)
    y_vals = np.arange(0, 10, 0.1)
    xx, yy = np.meshgrid(x_vals, y_vals)
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    return xx, yy, grid_points


def plot_decision_boundary(model, X_train, y_train):
    xx, yy, grid_points = generate_test_grid()
    predictions = model.predict(grid_points)
    
    plt.scatter(grid_points[predictions==0][:,0], grid_points[predictions==0][:,1], 
                s=5, alpha=0.3, color='blue')
    plt.scatter(grid_points[predictions==1][:,0], grid_points[predictions==1][:,1], 
                s=5, alpha=0.3, color='red')
    
    plt.scatter(X_train[:,0], X_train[:,1], c=y_train, edgecolors='black')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("kNN Decision Boundary")
    plt.show()
```

## Hypeparameter Tuning (A7)

```python
def tune_knn(X_train, y_train):
    param_grid = {'n_neighbors': np.arange(1, 31)}
    
    knn = KNeighborsClassifier()
    grid = GridSearchCV(knn, param_grid, cv=5, scoring='f1')
    grid.fit(X_train, y_train)
    
    return grid.best_params_, grid.best_score_
```

# MAIN function

```python
# Load dataset
data = load_wine_data("dataset/winequality-red.csv")   # change filename if white wine

# Create binary classes
data = create_class_labels(data)

features = data.drop(['quality', 'class'], axis=1)
labels = data['class']

X_train, X_test, y_train, y_test = split_and_scale_features(features, labels)

# Train kNN (k=3)
knn_model = train_knn_model(X_train, y_train, k=3)

# ---- A1 METRICS ----
train_cm, train_p, train_r, train_f1 = classification_metrics(knn_model, X_train, y_train)
test_cm, test_p, test_r, test_f1 = classification_metrics(knn_model, X_test, y_test)

print("TRAIN CONFUSION MATRIX:\n", train_cm)
print("TRAIN Precision:", train_p, "Recall:", train_r, "F1:", train_f1)

print("\nTEST CONFUSION MATRIX:\n", test_cm)
print("TEST Precision:", test_p, "Recall:", test_r, "F1:", test_f1)

# ---- A3 ----
X_rand, y_rand = generate_training_points()
plot_training_data(X_rand, y_rand)

# ---- A4 ----
knn_2d = train_knn_model(X_rand, y_rand, k=3)
plot_decision_boundary(knn_2d, X_rand, y_rand)

# ---- A5 ----
for k in [1, 3, 7, 15]:
    print(f"\nDecision boundary for k = {k}")
    model_k = train_knn_model(X_rand, y_rand, k)
    plot_decision_boundary(model_k, X_rand, y_rand)

# ---- A7 ----
best_k, best_score = tune_knn(X_train, y_train)

print("\nBest k from GridSearch:", best_k)
print("Best CV F1-score:", best_score)
```
